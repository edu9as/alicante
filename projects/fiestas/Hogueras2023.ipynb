## Descubre la magia de las Hogueras de Alicante: Un plan de visita optimizado usando K-Means

En esta sección, presentamos un enfoque innovador para planificar tus visitas a las Hogueras de Alicante de manera eficiente y efectiva. Utilizando el algoritmo de agrupamiento K-Means y la ubicación geográfica de las hogueras, hemos dividido estas en 4 o 5 clusters. Este enfoque nos permite optimizar tu tiempo y asegurarnos de que puedas disfrutar de todas las hogueras principales en un período de 4 o 5 días.

Mediante la técnica de K-Means, agrupamos las hogueras en clusters basados en su proximidad geográfica, lo que nos permite establecer rutas lógicas y eficientes para visitarlas. Cada cluster representa una zona geográfica con un conjunto de hogueras cercanas entre sí. De esta manera, puedes explorar y disfrutar de las Hogueras sin perder tiempo en desplazamientos innecesarios.

A través de esta planificación óptima, podrás sumergirte en la esencia de cada hoguera y aprovechar al máximo tu experiencia en las fiestas de Alicante. ¡Prepárate para disfrutar de una experiencia inolvidable con la ayuda de nuestro enfoque basado en K-Means para la planificación de visitas a las Hogueras de Alicante!

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

km5 = KMeans(n_clusters=5, random_state=42).fit(df[["lat", "lon"]])
km4 = KMeans(n_clusters=4, random_state=42).fit(df[["lat", "lon"]])

df["km4"] = [list(colors.values())[l] for l in km4.labels_]
df["km5"] = [list(colors.values())[l] for l in km5.labels_]

fig,ax = plt.subplots(1,2, figsize = (10,5))

df.plot.scatter(title="Todas las hogueras en 4 días", x="lon", y="lat", c = df.km4, ec="#000", ax = ax[0])
df.plot.scatter(title="Todas las hogueras en 5 días", x="lon", y="lat", c = df.km5, ec="#000", ax = ax[1])


ax[1].set_yticks([])
ax[1].set_ylabel("")

fig.tight_layout()

plt.show()
```

## Descubre la ruta perfecta para visitar todas las Hogueras de Alicante en 4 o 5 días: Optimización de rutas mediante algoritmo greedy

En esta sección, presentamos una solución innovadora para que puedas disfrutar de todas las Hogueras de Alicante en tan solo 4 o 5 días. Utilizando el algoritmo de agrupamiento K-Means para dividir las hogueras en clusters y un algoritmo greedy para optimizar las rutas, hemos diseñado un recorrido eficiente que te permitirá explorar cada una de las hogueras sin perder tiempo valioso.

El enfoque combina la potencia de K-Means para agrupar las hogueras en clusters basados en su proximidad geográfica y un algoritmo *greedy* que selecciona la hoguera más cercana en cada etapa del recorrido. Comenzando desde la hoguera más cercana al transporte público, te dirigirás de hoguera en hoguera, asegurándote de visitar cada una sin duplicar tus pasos.

Esta estrategia de optimización de rutas te permitirá aprovechar al máximo tu tiempo y disfrutar plenamente de las Hogueras de Alicante. Podrás sumergirte en la magia de cada hoguera, explorar sus detalles artísticos y sumarte a la celebración sin preocuparte por la planificación logística.

```python
import numpy as np
from geopy.distance import geodesic

def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).meters

mat = np.zeros((len(df), len(df)))

for i, r in df.iterrows():
    for j, q in df.iterrows():
        mat[i,j] = calculate_distance((r.lat, r.lon), (q.lat, q.lon))
```

```python
fogueres_and_distances = {
   "km4": [],
   "km5": [],
}

for km in fogueres_and_distances:
    for i,c in enumerate(df[km].unique()):
        df_ = df[df[km] == c]

        seen = []
        distances = []

        while len(df_) > len(seen):
            if not seen:
              im_at = df_.index.tolist()[df_.lat.argmin()]
              seen.append(im_at)
              distances.append(0)
            else:
              im_at = seen[-1]
            
            go_to = sorted(df_.index, key= lambda x: mat[im_at, x] if x not in seen else 1e7)[0]
            seen.append(go_to)
            distances.append(mat[im_at, go_to])

        fogueres_and_distances[km].append([seen,distances])
```
